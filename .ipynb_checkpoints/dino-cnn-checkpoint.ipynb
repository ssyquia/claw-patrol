```python
import os
import sys
import warnings
warnings.filterwarnings('ignore')
sys.stderr = open(os.devnull, 'w')
```

```python
%matplotlib inline
from matplotlib import pyplot as plt

import matplotlib.pyplot as plt 
import pandas as pd
import os
import random
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

## Prepare Data

```python
base_dir = "data"

batch_size = 32
target_image_size = (224, 224)
```

```python
# Image data augmentation settings
data_gen = tf.keras.preprocessing.image.ImageDataGenerator(
    rescale=1.0 / 255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode="nearest",
    validation_split=0.2,
)

# Train data generator
train_generator = data_gen.flow_from_directory(
    directory=base_dir,
    target_size=target_image_size,
    batch_size=batch_size,
    class_mode="categorical",
    subset="training",
)

# Test data generator
test_generator = data_gen.flow_from_directory(
    directory=base_dir,
    target_size=target_image_size,
    batch_size=batch_size,
    class_mode="categorical",
    subset="validation",
)
```

## Train the Model

```python
# Load the base model
base_model = tf.keras.applications.ResNet152V2(
    weights="imagenet", include_top=False, input_shape=target_image_size + (3,)
)

# Define custom classification layers
x = base_model.output
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dropout(0.5)(x)
x = tf.keras.layers.Dense(512, activation="relu")(x)

# Final classification layer
predictions = tf.keras.layers.Dense(15, activation="softmax")(x)

# Combine base model with custom classification layers
model = tf.keras.Model(inputs=base_model.input, outputs=predictions)

# Summary of model architecture
print("Number of base model layers: ", len(base_model.layers))

# Freeze base model layers
for layer in base_model.layers[:530]:
    layer.trainable = False

# Compile the model with a lower learning rate
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss="categorical_crossentropy",
    metrics=["accuracy"],
)

# Train the model
history = model.fit(
    train_generator,
    steps_per_epoch=len(train_generator),
    epochs=30,
    validation_data=test_generator,
    validation_steps=len(test_generator),
)
```

## Plot Performance

```python
epoch_vals = range(1, 30, 2)  # Adjusted to include only odd epochs

plt.figure(figsize=(10, 10))

plt.plot(epoch_vals, history.history['accuracy'][::2], label='Training Accuracy')
plt.plot(epoch_vals, history.history['val_accuracy'][::2], label='Validation Accuracy')
plt.title('Accuracy', fontsize=24)
plt.legend(loc='lower right');
```

```python
epoch_vals = range(1, 30, 2)  # Adjusted to include only odd epochs

plt.figure(figsize=(10, 10))

plt.plot(epoch_vals, history.history['loss'][::2], label='Training Loss')
plt.plot(epoch_vals, history.history['val_loss'][::2], label='Validation Loss')
plt.title('Accuracy', fontsize=24)
plt.legend(loc='lower right');
```

## Save the Model

```python
model.save('model/dino_classifier.keras')
```
